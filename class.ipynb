```python
import numpy as np
import pandas as pd
from skimpy import skim
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
id_test=test['id'].values
```

```python
skim(train)
```

```python
train.rename(columns={'Daytime/evening attendance': 'Morning/Evening', "Curricular units 1st sem (credited)": '1_sem(cred)',"Curricular units 1st sem (enrolled)":"1_sem(enr)","Curricular units 1st sem (evaluations)":"1_sem(eval)","Curricular units 1st sem (approved)":"1_sem(appr)","Curricular units 1st sem (grade)":"1_sem(garde)","Curricular units 2nd sem (credited)": '2_sem(cred)',"Curricular units 2nd sem (enrolled)":"2_sem(enr)","Curricular units 2nd sem (evaluations)":"2_sem(eval)","Curricular units 2nd sem (approved)":"2_sem(appr)","Curricular units 2nd sem (grade)":"2_sem(garde)"}, inplace=True)

test.rename(columns={'Daytime/evening attendance': 'Morning/Evening', "Curricular units 1st sem (credited)": '1_sem(cred)',"Curricular units 1st sem (enrolled)":"1_sem(enr)","Curricular units 1st sem (evaluations)":"1_sem(eval)","Curricular units 1st sem (approved)":"1_sem(appr)","Curricular units 1st sem (grade)":"1_sem(garde)","Curricular units 2nd sem (credited)": '2_sem(cred)',"Curricular units 2nd sem (enrolled)":"2_sem(enr)","Curricular units 2nd sem (evaluations)":"2_sem(eval)","Curricular units 2nd sem (approved)":"2_sem(appr)","Curricular units 2nd sem (grade)":"2_sem(garde)"}, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train['Target_'] = label_encoder.fit_transform(train['Target'])
```

```python
train.drop(columns=['id','Course','Marital status','Nacionality',"Curricular units 1st sem (without evaluations)","Curricular units 2nd sem (without evaluations)","Mother's qualification",'Previous qualification',"Mother's occupation","Father's qualification","Father's occupation",'Target'],inplace=True)

test.drop(columns=['id','Course','Marital status','Nacionality',"Curricular units 1st sem (without evaluations)","Curricular units 2nd sem (without evaluations)","Mother's qualification",'Previous qualification',"Mother's occupation","Father's qualification","Father's occupation"],inplace=True)
```

```python
train
```

```python
X=train.drop(columns=['Target_'])
y=train['Target_']
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.2) 
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```

## Neural Network

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```

## using Optuna

```python
import optuna
```

```python
from sklearn.ensemble import RandomForestClassifier
def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 1, 30)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        random_state=42
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    return accuracy_score(y_test, y_pred)

# Create a study and optimize
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10)#increase the n_traials for more iterations

# Best parameters
print("Best parameters: ", study.best_params)
print("Best accuracy: ", study.best_value)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=101, max_depth=18,min_samples_split=7)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
print(accuracy_score(y_test, y_pred)*100)
print(mean_squared_error(y_test, y_pred)*100)
print(mean_absolute_error(y_test, y_pred)*100)
```
